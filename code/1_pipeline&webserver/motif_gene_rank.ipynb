```python
df_p_min=pd.read_csv(dir+"motif_gene_p_find.txt",header=0,sep="\t",index_col=0)
```

```python
start_order=df_p_min.index.tolist()

df_rank=pd.DataFrame()

for col0 in df_p_min.columns:
    df_test_sort_b=df_p_min.sort_values(by=col0,ascending=True)
    ind_index=OrderedDict({})

    i=1
    for ww in df_test_sort_b.index:
        ind_index[ww]=i
        i=i+1
    rank_0=[]
    for ww in start_order:
        rank_0.append(ind_index[ww])
    df_rank[col0]=rank_0
df_rank.index=start_order

df_rank.T.to_csv(dir+"motif_gene_rank.txt",sep="\t",header=True,index=True)
```
